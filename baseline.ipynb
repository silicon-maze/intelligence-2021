```python
import pandas as pd
```

# Import Datasets

```python
train_data = pd.read_csv("train_dataset.csv")
test_data = pd.read_csv("test_dataset.csv")
```

# Convert datatype of selected fields

```python
train_data["profession"] = pd.factorize(train_data.profession)[0]
train_data["home_city"] = pd.factorize(train_data.home_city)[0]
train_data["home_state"] = pd.factorize(train_data.home_state)[0]
train_data["has_child"] = pd.factorize(train_data.has_child)[0]
train_data["residential_status"] = pd.factorize(train_data.residential_status)[0]
train_data["owns_bike"] = pd.factorize(train_data.owns_bike)[0]
```

```python
test_data["profession"] = pd.factorize(test_data.profession)[0]
test_data["home_city"] = pd.factorize(test_data.home_city)[0]
test_data["home_state"] = pd.factorize(test_data.home_state)[0]
test_data["has_child"] = pd.factorize(test_data.has_child)[0]
test_data["residential_status"] = pd.factorize(test_data.residential_status)[0]
test_data["owns_bike"] = pd.factorize(test_data.owns_bike)[0]
```

# Drop the null and dependent variable from the train dataset. 

```python
train_data.dropna(inplace=True)

train_X = train_data.drop("ineligible_for_funds",axis=1)
train_y = train_data["ineligible_for_funds"]
```

# Train the Baseline model

```python
from sklearn.tree import DecisionTreeClassifier
DTClassifier= DecisionTreeClassifier()
DTClassifier.fit(train_X, train_y)
```

# Predict on Test data set

```python
y_pred = pd.DataFrame(DTClassifier.predict(test_data))
```

```python
submission = pd.concat([test_data.Id, y_pred], axis=1)
submission.to_csv('submission.csv', index=False)
```
